```python
n = 16
def power_of_two(n: int) -> bool:
    if n <=0:
        return False,None
    power = 0
    while n%2 == 0:
        n = n // 2
        power = power + 1
    if n == 1:
        return True,power
    else:
        return False, None
if __name__=="__main__":
    try:
        n = 16
        power_of_two()
    except Exception as e:
        print(e)
```
